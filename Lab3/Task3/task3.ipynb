```python
import numpy as np
from scipy.spatial.distance import euclidean
from fastdtw import fastdtw

# Generate two sine waves with different frequencies
freq1 = 1  # Frequency of the first sine wave
freq2 = 2  # Frequency of the second sine wave
length = 100  # Length of the sequences

# Generate the temporal sequences (sine waves)
sequence1 = np.sin(np.linspace(0, 2*np.pi*freq1, length))
sequence2 = np.sin(np.linspace(0, 2*np.pi*freq2, length))

# Calculate Euclidean distance
euclidean_distance = euclidean(sequence1, sequence2)

# Calculate DTW distance
dtw_distance, _ = fastdtw(sequence1, sequence2)

# Print the distances
print("Euclidean Distance:", euclidean_distance)
print("DTW Distance:", dtw_distance)
```

```python
import matplotlib.pyplot as plt

# Plot the sequences
plt.figure(figsize=(10, 6))
plt.plot(sequence1, label='Sequence 1 (Freq = 1)')
plt.plot(sequence2, label='Sequence 2 (Freq = 2)')
plt.title('Temporal Sequences')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```
