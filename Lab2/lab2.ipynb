```python
from statsmodels.tsa.arima_process import ArmaProcess
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller

plt.rcParams['figure.figsize'] = (10, 5)
ar_1 = np.array([1, 0.8])
ma_1 = np.array([1]) # MA part is not present
ar_proc_1 = ArmaProcess(ar_1, [ma_1])
ar_data_1 = ar_proc_1.generate_sample(nsample=1000)
plt.plot(ar_data_1)
plt.savefig('plots/1_1_line.png')
print(ar_proc_1.isstationary)
result_1 = adfuller(ar_data_1, autolag='AIC')
print('ADF Statistic: %f' % result_1[0])
print('p-value: %f' % result_1[1])
print('Critical Values:')
for key, value in result_1[4].items():
    print('\t%s: %.3f' % (key, value))
```

Task 2

```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima_process import ArmaProcess
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def generate_and_plot(ar, ma, plot_filename_base):
    ar_proc = ArmaProcess(ar, ma)
    ar_data = ar_proc.generate_sample(nsample=1000)
    series = pd.Series(ar_data)

    # Line plot
    plt.figure()
    series.plot.line()
    plt.savefig(f'plots/{plot_filename_base}_line.png')

    # Histogram
    plt.figure()
    series.plot.hist()
    plt.savefig(f'plots/{plot_filename_base}_hist.png')

    # Density plot
    plt.figure()
    series.plot.density()
    plt.savefig(f'plots/{plot_filename_base}_density.png')

    # Box plot
    plt.figure()
    series.plot.box()
    plt.savefig(f'plots/{plot_filename_base}_box.png')

    # Lag-1 plot
    plt.figure()
    pd.plotting.lag_plot(series, lag=1)
    plt.savefig(f'plots/{plot_filename_base}_lag1.png')

    # Lag-2 plot
    plt.figure()
    pd.plotting.lag_plot(series, lag=2)
    plt.savefig(f'plots/{plot_filename_base}_lag2.png')

    # ACF graph
    plt.figure()
    plot_acf(series)
    plt.savefig(f'plots/{plot_filename_base}_acf.png')

    # PACF graph
    plt.figure()
    plot_pacf(series)
    plt.savefig(f'plots/{plot_filename_base}_pacf.png')


    print('stationary: ' + str(ar_proc.isstationary))
    print('invertible: ' + str(ar_proc.isinvertible))
    result = adfuller(ar_data, autolag='AIC')
    print('ADF Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
```

```python
ar_1 = np.array([1, 0.8])
ma_1 = np.array([1]) # MA part is not present
generate_and_plot(ar_1, ma_1, '2_1')
```

```python
ar_1 = np.array([1, -0.8])
ma_1 = np.array([1]) # MA part is not present
generate_and_plot(ar_1, ma_1, '2_2')
```

```python
ar_1 = np.array([1, 1, -0.5])
ma_1 = np.array([1]) # MA part is not present
generate_and_plot(ar_1, ma_1, '2_3')
```

```python
ar_1 = np.array([1, -1, -0.5])
ma_1 = np.array([1]) # MA part is not present
generate_and_plot(ar_1, ma_1, '2_4')
```
